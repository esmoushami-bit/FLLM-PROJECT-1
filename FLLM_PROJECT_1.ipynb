```python
import numpy as np

# Sample text
text = "I love learning AI with transformers"

# 1. Tokenization
tokens = text.lower().split()
print("Tokens:", tokens)

# 2. Vocabulary
vocab = {word:i for i,word in enumerate(sorted(set(tokens)))}
token_ids = [vocab[t] for t in tokens]
print("Token IDs:", token_ids)

# 3. Embedding Layer
embed_dim = 8
embedding = np.random.rand(len(vocab), embed_dim)
embeddings = embedding[token_ids]
print("Embedding Shape:", embeddings.shape)

# 4. Batch and Multi-head preparation
batch = 1
heads = 2
seq_len = len(token_ids)
d = embed_dim // heads
x = embeddings.reshape(batch, seq_len, heads, d).transpose(0,2,1,3)

# 5. Query, Key, Value
Q = x.copy()
K = x.copy()
V = x.copy()

# 6. Scaled Dot-Product Attention
scores = np.matmul(Q, K.transpose(0,1,3,2)) / np.sqrt(d)

# 7. Causal Mask
mask = np.triu(np.ones((seq_len, seq_len)), 1).astype(bool)
scores = np.where(mask, -1e9, scores)

# 8. Softmax
scores = scores - np.max(scores, axis=-1, keepdims=True)
weights = np.exp(scores)
weights /= np.sum(weights, axis=-1, keepdims=True)

# 9. Attention Output
output = np.matmul(weights, V)
output = output.transpose(0,2,1,3).reshape(batch, seq_len, embed_dim)

print("Attention Output Shape:", output.shape)
print(output)
```
